```python
import pandas as pd

lego = pd.read_csv('lego_sets.csv')

lego.head()

theme = pd.read_csv('parent_themes.csv')

### outputs by default 5 rows
theme.head()
```

```python
lego.head()
```

```python
merged_df = lego.merge(theme, left_on ='parent_theme', right_on = 'name')

merged_df.drop(columns = 'name_y', inplace = True)

merged_df.head()
```

```python
##### The shape (rows, columns) of the filtered set_nums column removing all null rows in the merged_df data frame
merged_df[merged_df['set_num'].isnull()].shape
```

```python
## Shows all licencesed star wars theme names 
licensed_rws = merged_df[merged_df['is_licensed']]
## drop na drops any rows that have null values like "NaN"
licensed_rws = licensed_rws.dropna(subset= ['set_num'])
licensed_rws.head()

#### What percentage of themes are star wars ####
#### of the licensed rows which ones have a parent theme that is == to Star Wars

star_wars = licensed_rws[licensed_rws['parent_theme']=='Star Wars']
# star_wars.head(50)

#### Could way to see the length of a data frame (rows, columns)
# star_wars.shape

### What percentage of all licencesed sets ever released were Star Wars themed
the_force = int(star_wars.shape[0]/licensed_rws.shape[0]*100)

print(f"{the_force}%")
```

```python
license_sorted = licensed_rws.sort_values('year')
license_sorted['count'] = 1

sum_df= license_sorted.groupby(['year','parent_theme']).sum().reset_index()
# sum_df.head()

sum_df.sort_values('count').drop_duplicates(['year'])

max_df = sum_df.sort_values('count', ascending = False).drop_duplicates(['year'])

max_df.sort_values('year', inplace = True)
max_df

## The year star wars got beat out
#### new_era = 2017
```

```python
#### BREAK DOWN THE NUMBER OF SETS BY YEAR ######

clean_df = merged_df [~merged_df['set_num'].isnull()]
### the ~ is for the opposite of what comes after it - not is null ###

#clean_df.head()

clean_df['count'] = 1
sets_by_year = clean_df.groupby(['year']).sum().reset_index()[['year','count']]
### only want to return year and count column in order of the year (groupby)

sets_by_year.head(50)
```
